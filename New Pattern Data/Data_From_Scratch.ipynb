```python
import numpy as np
from numpy.random import normal as randn
import pandas as pd
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python


T = 8

def pattern_data(mu, sigma, scale_size, time_point, pattern_type, w):

    for i in range(time_point):
        #parameter
        P = np.arange(17,time_point, 1)
        s = np.arange(1.5*sigma, 2.5*sigma, (2.5*sigma-1.5*sigma)/time_point)
        k = np.zeros(time_point)
        k1 = np.put(k, P, 1)

        if pattern_type ==1:
            y = mu + randn(0,scale_size,[w,time_point])*sigma #normal

        elif pattern_type ==2:
            y = mu + randn(0,scale_size,[w,time_point])*sigma + np.arange(1, time_point+1, 1)*(np.arange(0.05*sigma, 0.1*sigma, 0.25/time_point)+2) #uptrend

        elif pattern_type ==3:
            y = mu + randn(0,scale_size,[w,time_point])*sigma - np.arange(1, time_point+1, 1)*(np.arange(0.05*sigma, 0.1*sigma, 0.25/time_point)+2) #downtrend

        elif pattern_type ==4:
            y = mu + randn(0,scale_size,[w,time_point]) * np.arange(0.2*sigma, 0.4*sigma, (0.4*sigma-0.2*sigma)/time_point) #stratification

        elif pattern_type ==5:
            y = mu + randn(0,scale_size,[w,time_point])*sigma + (np.arange(1*sigma, 3*sigma, (3*sigma - 1*sigma)/time_point)+1)*((-1)**np.arange(1, time_point+1, 1)) #systematic

        elif pattern_type ==6:
            y = mu + randn(0,scale_size,[w,time_point])*sigma + np.arange(1*sigma, 3*sigma, (3*sigma-1*sigma)/time_point)*np.sin((2*np.pi*np.arange(1 ,time_point+1 ,1))/T) #cyclic

        elif pattern_type ==7:
            y = mu + randn(0,scale_size,[w,time_point])*sigma + k*s #upshift

        elif pattern_type ==8:
            y = mu + randn(0,scale_size,[w,time_point])*sigma - k*s #downshift


    return y
```

```python
#normal_pattern

y1 = pattern_data(80, 5, 1, 32, 1, 500)
y1_data = pd.DataFrame(y1)
y1_data.to_csv(r"C:\Users\user\Desktop\SEM 8\Project Sarjana Muda (PSM)\THESIS PROGRESS\Control Chart Pattern Recognition (CCPR) Project\New Pattern Data\Normal.csv")

plt.figure(figsize=(10,5))
plt.axhline(80, color='black')
plt.plot(y1_data.iloc[1,:])
```
